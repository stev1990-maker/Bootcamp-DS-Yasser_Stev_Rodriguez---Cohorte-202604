# Introducción a las apis


## ¿Qué son las APIs y cómo funcionan?

---

Una API (Application Programming Interface) o Interfaz de Programación de Aplicaciones es esencialmente un intermediario.
Permite que dos aplicaciones de software se comuniquen entre sí. Imagina un camarero en un restaurante:
tú (una aplicación) le dices al camarero (la API) lo que quieres pedir (una solicitud),
el camarero va a la cocina (otro sistema) y trae tu comida (la respuesta).

En el contexto de las APIs, esto se traduce en:

1.  **Solicitud (Request):** Una aplicación envía una solicitud a otra aplicación a través de su API. Esta solicitud especifica qué acción se desea realizar (por ejemplo, obtener datos, enviar información).
2.  **Respuesta (Response):** La aplicación que recibe la solicitud procesa la petición y envía una respuesta de vuelta a la aplicación solicitante. Esta respuesta contiene los datos solicitados o un mensaje indicando el resultado de la acción.

Existen diferentes estilos o arquitecturas para diseñar APIs, siendo **REST (Representational State Transfer)** uno de los más comunes. Las APIs RESTful utilizan métodos estándar de HTTP (como GET para obtener datos, POST para enviar datos, PUT para actualizar y DELETE para eliminar) para interactuar con recursos.

## Ejemplos de APIs en la vida real:
---

*   Cuando inicias sesión en un sitio web usando tu cuenta de Google o Facebook, estás utilizando sus APIs de autenticación.
*   Las aplicaciones del clima obtienen los datos meteorológicos de APIs proporcionadas por servicios meteorológicos.
*   Las aplicaciones de mapas utilizan APIs para obtener información de ubicación, rutas y puntos de interés.

## Códigos de estado HTTP comunes que indican errores:
---
Los códigos de estado HTTP son números de tres dígitos que el servidor envía en la respuesta para indicar el resultado de la solicitud. Los códigos en el rango 4xx y 5xx generalmente indican errores:

- **400 Bad Request:** La solicitud no pudo ser entendida por el servidor debido a una sintaxis inválida.
- **401 Unauthorized:** La solicitud requiere autenticación. Puede que no hayas proporcionado credenciales o que estas sean inválidas.
- **403 Forbidden:** La solicitud fue válida, pero el servidor se rehúsa a autorizarla. Puede que no tengas permisos para acceder al recurso.
- **404 Not Found:** El servidor no pudo encontrar el recurso solicitado. El URL podría estar incorrecto.
- **500 Internal Server Error:** El servidor encontró una condición inesperada que le impidió completar la solicitud. Esto es un error general del lado del servidor.
- **503 Service Unavailable:** El servidor no está listo para manejar la solicitud. Esto puede ocurrir si el servidor está sobrecargado o en mantenimiento.

Cómo verificar el `response.status_code`:

Después de realizar una solicitud utilizando la librería `requests`, el objeto `response` tiene un atributo `status_code` que contiene el código de estado HTTP. Podemos usar una sentencia `if` para verificar este código.

```python
import requests

base_url = "https://api-colombia.com/api/v1"
endpoint = "/Department" # Este endpoint debería funcionar (código 200)
invalid_endpoint = "/NonExistentEndpoint" # Este endpoint debería generar un error 404

response_valid = requests.get(base_url + endpoint)
print(f"\nSolicitud a {base_url + endpoint}")
print(f"Código de estado: {response_valid.status_code}")

if response_valid.status_code == 200:
    print("Solicitud exitosa.")
elif response_valid.status_code == 404:
    print("Error 404: Recurso no encontrado.")
elif response_valid.status_code == 401:
    print("Error 401: No autorizado.")
else:
    print(f"Error inesperado: Código de estado {response_valid.status_code}")
```

```python
response_invalid = requests.get(base_url + invalid_endpoint)
print(f"\nSolicitud a {base_url + invalid_endpoint}")
print(f"Código de estado: {response_invalid.status_code}")

if response_invalid.status_code == 200:
    print("Solicitud exitosa.")
elif response_invalid.status_code == 404:
    print("Error 404: Recurso no encontrado.")
    print("Verifica que el URL del endpoint sea correcto.")
elif response_invalid.status_code == 401:
    print("Error 401: No autorizado.")
else:
    print(f"Error inesperado: Código de estado {response_invalid.status_code}")
```

## Beneficios de usar APIs:
---
**Para desarrolladores:**
*   Ahorro de tiempo y esfuerzo al no tener que crear funcionalidades desde cero.
*   Acceso a datos y servicios de terceros para enriquecer sus propias aplicaciones.
*   Mayor modularidad y mantenibilidad del código.

**Para usuarios:**
*   Experiencias de usuario más integradas y fluidas (por ejemplo, iniciar sesión con una sola cuenta).
*   Acceso a una mayor variedad de funcionalidades y servicios dentro de una sola aplicación.
