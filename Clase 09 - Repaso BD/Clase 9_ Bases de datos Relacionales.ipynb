# Crear una tabla

```python
import pyodbc

# Parámetros de conexión
server = 'cdsc-bit.database.windows.net'
database = 'BDPruebasBIT'
username = 'DDLUser1'
password = 'Hola2025'
driver = '{ODBC Driver 17 for SQL Server}'
conn = pyodbc.connect(
    f'DRIVER={driver};SERVER={server};DATABASE={database};UID={username};PWD={password};Encrypt=yes;TrustServerCertificate=no;Connection Timeout=30;'
)
cur = conn.cursor()
```

```python
cur.execute('''
CREATE TABLE alumnos (
    id INTEGER PRIMARY KEY IDENTITY(1,1),
    nombre VARCHAR(200),
    edad INTEGER,
    ciudad VARCHAR(200)
)
''')
```

```python
cur.execute('''
DROP TABLE alumnos
''')
```

```python
conn.commit()
```

## Ingresar datos

```python
import pyodbc

# Parámetros de conexión
server = 'cdsc-bit.database.windows.net'
database = 'BDPruebasBIT'
username = 'WriterUser1'
password = 'Hola2025'
driver = '{ODBC Driver 17 for SQL Server}'
conn = pyodbc.connect(
    f'DRIVER={driver};SERVER={server};DATABASE={database};UID={username};PWD={password};Encrypt=yes;TrustServerCertificate=no;Connection Timeout=30;'
)
cur = conn.cursor()
```

```python
alumnos = [
    ('Ana', 22, 'Bogotá'),
    ('Luis', 24, 'Medellín'),
    ('Marta', 23, 'Cali'),
    ('Pedro', 22, 'Bogotá')
]

cur.executemany('INSERT INTO alumnos VALUES ( ?, ?, ?)', alumnos)
conn.commit()
```

```python
alumnos = [
    ('Cristian', 27, 'Bogotá'),
]

cur.executemany('INSERT INTO alumnos VALUES ( ?, ?, ?)', alumnos)
conn.commit()
```

```python
cur.execute('''

BEGIN TRANSACTION

UPDATE alumnos
SET edad = edad + 1
WHERE id = 5

ROLLBACK TRANSACTION

''')
```

```python
conn.commit()
```

```python
cur.execute('''
DELETE alumnos
''')
```

```python
conn.rollback()  ## Regresa a la ultima version (ultimo commit)
```

```python
pd.read_sql_query("SELECT * FROM alumnos", conn)
```

```python
# Consulta todos los registros
cur.execute('SELECT * FROM alumnos')
for row in cur.fetchall():
    print(row)
```

```python
conn.commit()
```

## Ingresar datos de un CSV

```python
formulario = pd.read_csv('/content/Asistencia.csv')
formulario
```

```python
import pyodbc

# Parámetros de conexión
server = 'cdsc-bit.database.windows.net'
database = 'BDPruebasBIT'
username = 'WriterUser1'
password = 'Hola2025'
driver = '{ODBC Driver 17 for SQL Server}'
conn = pyodbc.connect(
    f'DRIVER={driver};SERVER={server};DATABASE={database};UID={username};PWD={password};Encrypt=yes;TrustServerCertificate=no;Connection Timeout=30;'
)
cur = conn.cursor()
```

```python
for _, row in formulario.iterrows():
    cur.execute('INSERT INTO alumnos (nombre, edad, ciudad) VALUES (?, ?, ?)', (row['Nombre'], row['Edad'], row['Ciudad']))
conn.commit()
```

```python
consulta = pd.read_sql_query("SELECT * FROM alumnos", conn)
consulta
```

```python
consulta.groupby('ciudad').count()
```

```python
pd.read_sql_query("SELECT ciudad, count(id) FROM alumnos group by ciudad", conn)
```

```python
consulta = pd.read_sql_query("SELECT * FROM alumnos", conn)
consulta
```

```python
cur.execute('''
DELETE alumnos
WHERE edad < 25
''')
```

```python
consulta = pd.read_sql_query("SELECT Ciudad, Count(*) FROM alumnos GROUP BY Ciudad HAVING count(*) = 5", conn)
consulta
```

```python
consulta = pd.read_sql_query("SELECT TOP 5 * FROM alumnos ORDER BY nombre", conn)
consulta
```
