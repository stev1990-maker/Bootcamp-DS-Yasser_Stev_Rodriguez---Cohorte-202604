### 2.1 Aprendizaje Supervisado
Se utiliza un conjunto de datos etiquetados para entrenar el modelo. (ej: clasificación, regresión)

**Ejemplo Práctico:** Clasificación de Flores Iris

```python
# Importar bibliotecas
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar datos
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

```python
X_test
```

```python
iris.target
```

```python
# Entrenar modelo
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
knn.predict(X_test)
```

```python
# Evaluar
print(f"Precisión: {knn.score(X_test, y_test):.2f}")
```

```python
X_train.shape,y_train.shape
```

```python
# Visualización con leyenda
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)

# Añadir leyenda usando los target_names
plt.legend(handles=scatter.legend_elements()[0],
           title="Especies",
           labels=iris.target_names.tolist())

plt.xlabel('Largo de sépalo (cm)')
plt.ylabel('Ancho de sépalo (cm)')
plt.title('Distribución de Flores Iris')
plt.show()
```

### 2.2 Aprendizaje No Supervisado
Se trabaja con datos no etiquetados y se busca descubrir patrones o estructuras subyacentes. (ej: clustering)

**Ejemplo Práctico:** Clustering con K-Means

```python
from sklearn.cluster import KMeans

# Crear clusters
kmeans = KMeans(n_clusters=3)
X = kmeans.fit(iris.data)
```

```python
kmeans.cluster_centers_
```

```python
# Visualización
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=X.labels_)
plt.plot(kmeans.cluster_centers_[0][0], kmeans.cluster_centers_[0][1], 'ro',markersize=16, alpha = 0.5)
plt.plot(kmeans.cluster_centers_[1][0], kmeans.cluster_centers_[1][1], 'bo',markersize=16, alpha = 0.5)
plt.plot(kmeans.cluster_centers_[2][0], kmeans.cluster_centers_[2][1], 'ko',markersize=16, alpha = 0.5)
plt.grid()
plt.xlabel('Largo de sépalo (cm)')
plt.ylabel('Ancho de sépalo (cm)')

# Añadir leyenda usando los target_names
plt.legend(handles=scatter.legend_elements()[0],
           title="Especies",
           labels=['virginica','setosa','versicolor'])

plt.xlabel('Largo de sépalo (cm)')
plt.ylabel('Ancho de sépalo (cm)')
plt.title('Distribución de Flores Iris')
plt.show()
```

## 3. Diferencias con Automatización Tradicional

- **Machine Learning:** Aprende patrones a partir de datos, lo que permite adaptarse y mejorar con el tiempo.
- **Automatización:** Sigue reglas y procedimientos predefinidos sin capacidad de adaptación.




 | Machine Learning | Automatización Tradicional |
 |------------------|----------------------------|
 | Aprende de datos | Sigue reglas predefinidas |
 | Adaptable        | Estático                  |
 | Maneja complejidad | Tareas estructuradas   |

**Ejemplo Comparativo:** Detección de Spam

**Automático:** Reglas basadas en palabras clave



```python
def spam_detector(email):
    spam_words = ['oferta', 'dinero', 'gratis','rápido']
    return any(word in email.lower() for word in spam_words)
```

```python
spam_detector("¡Eres el ganador!")
```

**Machine Learning:** Clasificación aprendida de ejemplos

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import pandas as pd
```

```python
# Datos de ejemplo
emails = ["Gana dinero rapido", "Reunion importante mañana", "Oferta exclusiva para ti"]
labels = [1, 0, 1]  # 1 = spam, 0 = no spam

labels_df = pd.DataFrame(labels,columns=['Es Spam?'])
```

```python
# Preprocesamiento
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)
X_df = pd.DataFrame(X.todense(), columns = vectorizer.get_feature_names_out())
pd.concat([X_df,labels_df],axis=1)
```

```python
# Entrenar modelo
clf = MultinomialNB()
clf.fit(X, labels)
```

```python
# Predecir
test_email = ["¡Eres el ganador!"]
print("Predicción:", clf.predict(vectorizer.transform(test_email))[0])
```

**Demo Rápida:** Reconocimiento de Dígitos (MNIST)


```python
from sklearn.datasets import load_digits
from sklearn.svm import SVC
```

```python
digits = load_digits()
X, y = digits.data, digits.target
```

```python
model = SVC()
model.fit(X[:10], y[:10])  # Usar subset para velocidad
```

```python
# Predicción de ejemplo
import matplotlib.pyplot as plt
test = 1796
plt.imshow(digits.images[test], cmap='gray')
digits.target_names[digits.target[test]]
```

```python
print("Predicción:", model.predict([digits.data[test]])[0])
plt.show()
```
